```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(
    r"C:\Users\hudso\OneDrive\Desktop\Data_Visualization_And_Modeling_Online-main\Data_Visualization_And_Modeling_Online-main\Lab\titanic_passengers.csv"
)

# Fill missing ages with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Create Age categories using .loc
df['Age_group'] = 'Unknown'

df.loc[df['Age'] < 18, 'Age_group'] = 'Child'
df.loc[(df['Age'] >= 18) & (df['Age'] < 60), 'Age_group'] = 'Adult'
df.loc[df['Age'] >= 60, 'Age_group'] = 'Senior'

# Visualization
sns.countplot(x='Age_group', hue='Survived', data=df)

plt.title('Survival Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset using full pathway
df = pd.read_csv(
    r"C:\Users\hudso\OneDrive\Desktop\Data_Visualization_And_Modeling_Online-main\Data_Visualization_And_Modeling_Online-main\Lab\titanic_passengers.csv"
)

# Fill missing Age values correctly (no warning)
df["Age"] = df["Age"].fillna(df["Age"].median())

# Create Age Groups
df["Age_group"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 35, 60, 100],
    labels=["Child", "Teen", "Young Adult", "Adult", "Senior"]
)

# Create crosstab for survival by age group
age_survival = pd.crosstab(df["Age_group"], df["Survived"])

# Plot
age_survival.plot(kind="bar")

plt.title("Survival Count by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])

plt.show()
```
